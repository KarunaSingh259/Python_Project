```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv ("Expanded_data_with_more_features.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

# Drop unnamed column

```python
df = df.drop("Unnamed: 0" , axis= 1)
print (df.head())
```

# **Gender Distribution**

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

From the above chart we have analysed that the number of females in the data is more than number of males.

# *Parent Education**

```python
gb = df.groupby("ParentEduc").agg({"MathScore": 'mean', "ReadingScore": 'mean' , "WritingScore": 'mean' })
print(gb)
```

```python
sns.heatmap(gb, annot= True)
plt.show()
```

From the abovr chart we have concluded that the education of the parents have a good impact on their childern education also.

# **Parent Marital Status**

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore": 'mean', "ReadingScore": 'mean' , "WritingScore": 'mean' })
print(gb1)
```

```python
sns.heatmap(gb1, annot= True)
plt.show()
```

From a above chart we analysis that there is no/negligible impact on the student score due to their parents marital status.

# **Practice Sport**

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data = df, x = "PracticeSport")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
gb2 = df.groupby("PracticeSport").agg({"MathScore": 'mean', "ReadingScore": 'mean' , "WritingScore": 'mean' })
print(gb2)
```

```python
sns.heatmap(gb2, annot= True)
plt.show()
```

From the above chart we can analysis that the student who practice sport regulary perform better in exam who never play sports and playing sometime.

## **Is First Child**

```python
gb3 = df.groupby("IsFirstChild").agg({"MathScore": 'mean', "ReadingScore": 'mean' , "WritingScore": 'mean' })
print(gb3)
```

```python
sns.heatmap(gb3, annot= True)
plt.show()
```

First children have a slight academic advantage in Math, Reading, and Writing compared to non-first children, but the difference is not significant enough to suggest a strong correlation.

# **Number of Siblings**

```python
gb4 = df.groupby("NrSiblings").agg({"MathScore": 'mean', "ReadingScore": 'mean' , "WritingScore": 'mean' })
print(gb4)
```

```python
sns.heatmap(gb4, annot= True)
plt.show()
```

The number of siblings does not strongly impact academic performance in Math, Reading, or Writing. The differences are very small and likely not statistically significant.

# **Weekly Study Hours**

```python
gb5 = df.groupby("WklyStudyHours").agg({"MathScore": 'mean', "ReadingScore": 'mean' , "WritingScore": 'mean' })
print(gb5)
```

```python
sns.heatmap(gb5, annot= True)
plt.show()
```

Students who study more than 10 hours per week consistently achieve the highest scores, while those studying less than 5 hours score the lowest across all subjects.
