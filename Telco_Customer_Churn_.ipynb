```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
df.head()
```

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
def conv(value):
  if value == 1:
    return "yes"
  else:
    return "no"

df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

```python
df.head()
```

converted 1 and 0 values of Senior Citizen to yes/no to make it easier to understand

```python
ax =   sns.countplot(x = 'Churn', data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Churn")
plt.show()
```

```python
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'],labels = gb.index, autopct = "%1.2f%%")
plt.title("Percentage of Churned Customeres")
plt.show()
```

From the give pie chart we can conclude that 26.54% of our customers have churned out .

```python
sns.countplot(x = 'gender', hue = 'Churn', data = df)
plt.title("Churn by Gender")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Crosstab (row-wise %)
ct = pd.crosstab(df['SeniorCitizen'], df['Churn'], normalize='index') * 100

# Plot stacked bar with custom colors
ax = ct.plot(kind='bar',
             stacked=True,
             figsize=(8,5),
             color=['#1f77b4', '#ff7f0e'])  # Blue & Orange

# Add percentage labels
for c in ax.containers:
    ax.bar_label(c, fmt='%.1f%%', label_type='center', color='white', fontsize=10)

plt.title("Churn by Senior Citizen (%)")

plt.ylabel("Percentage")
plt.legend(title="Churn")
plt.show()
```

```python
ax= sns.countplot(x = "SeniorCitizen",data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customer by Senior Citizens")
plt.show()
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

# List of categorical columns
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Custom palette for Churn (No = Blue, Yes = Orange)
churn_palette = {"No": "#1f77b4", "Yes": "#ff7f0e"}

# Create subplot grid (3 rows Ã— 3 columns for 9 plots)
fig, axes = plt.subplots(3, 3, figsize=(16, 12))
axes = axes.flatten()

# Loop through columns and plot with churn-based colors
for i, col in enumerate(cols):
    ax = sns.countplot(x=col, data=df, ax=axes[i], hue='Churn', palette=churn_palette)
    axes[i].set_title(f"Count of {col}", fontsize=12)
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Count")

    # Add labels on bars
    for container in ax.containers:
        ax.bar_label(container, fmt='%d', fontsize=9, color="black")

# Remove empty subplot if columns < grid size
for j in range(len(cols), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

The majority of Telco customers subscribe to basic services such as Phone Service and Internet Service (DSL/Fiber optic), while adoption of additional services like Online Security, Tech Support, and Device Protection remains comparatively low. Notably, churn rates are significantly higher among Fiber optic users and customers who do not opt for these value-added services. This indicates that retaining customers may require both improving the Fiber optic experience and encouraging adoption of supportive services.
